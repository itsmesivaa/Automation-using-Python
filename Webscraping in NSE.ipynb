```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
#Sample List of stocks
nse_stk = ['TCS','SBIN','WIPRO','MRF']
```

```python
nse_stk
```

```python
#NSE Base URL
base_nse_url = "https://www.nseindia.com/get-quotes/equity?symbol="
```

```python
base_nse_url
```

```python
#Generating custom NSE url to fetch data for respective stocks with stock name and url pair
x = []

for stk_name in range(0,len(nse_stk)):
    x.append({nse_stk[stk_name]:base_nse_url + nse_stk[stk_name]})
    #x.append(base_nse_url + nse_stk[stk_name])
    print(base_nse_url + nse_stk[stk_name])
```

```python
x
```

```python
#Generating response object through custom url built 
url = "https://www.nseindia.com/get-quotes/equity?symbol=WIPRO"
page = requests.get(url, headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36"})
print(page)
soup = BeautifulSoup(page.text,"html")
```
